```python
import csv
import matplotlib.pyplot as plt

valeur = []
first = []
second = []
third = []

with open('./complexite.csv', 'r') as file:
    reader = csv.reader(file)
    
    next(reader)
    
    for row in reader:
        valeur.append(float(row[0]))
        first.append(float(row[1]))
        second.append(float(row[2]))
        third.append(float(row[3]))
```

```python
import numpy as np
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

linwidth = 1
window_size = 100 

x = valeur[window_size-1:]
y_smooth = moving_average(first, window_size)
z_smooth = moving_average(second, window_size)
r_smooth = moving_average(third, window_size)

plt.figure(figsize=(10, 10))
plt.plot(x, y_smooth, linewidth=linwidth)
plt.plot(x, z_smooth, linewidth=linwidth)
plt.plot(x, r_smooth, linewidth=linwidth)

plt.legend(['O1', 'O2', 'O3'])
plt.show()
```
